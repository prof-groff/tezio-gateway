```python
from pynput.keyboard import Key, Listener
print('hello world')
from IPython.display import clear_output

def on_key_release(key):
    if key == Key.up:
        clear_output(wait=True)
        print('up pressed')
    if key == Key.down:
        clear_output(wait=True)
        print('down pressed')
    if key == Key.right:
        clear_output(wait=True)
        print('right pressed')
    if key == Key.left:
        clear_output(wait=True)
        print('left pressed')
    if key == Key.esc:
        clear_output(wait=True)
        print('escape')
        return False

with Listener(on_release = on_key_release) as listener:
    listener.join()
```
